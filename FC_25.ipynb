#### 2. Les gauchers sont réputés pour avoir un plus mauvais pied faible que les droitiers. Analysez les mesures de tendance centrale et de dispersion de la variable Weak Foot pour ces deux groupes. Comparez leurs distributions et formulez une hypothèse pour vérifier cette idée reçue.

```python
left_footed = df[df['Preferred foot'] == 0]['Weak foot']
right_footed = df[df['Preferred foot'] == 1]['Weak foot']

plt.figure(figsize=(10, 6))
sns.histplot(left_footed, kde=True, color='blue', label='Left')
sns.histplot(right_footed, kde=True, color='red', label='Right')
plt.title('Weak Foot Distribution')
plt.legend()
plt.show()
```

Les droitiers, en moyenne, maitrisent mieux le pied gauche que les gauchers maitrisent le droit.

Vu que les droitiers sont en plus grand nombre, il existe plus de compétences sur les pieds faibles.

Les droitiers ont une couverture de weak foot bien plus haute que les gauchers, donc une facilité à utiliser leurs pieds.

Cela confirme donc que les droitiers maitrisent mieux leur pied faible.

#### 3.	Parmi les compétences PAC, SHO, PAS, DRI et DEF, déterminez laquelle ou lesquelles semblent avoir le plus d'impact sur la note générale (OVR) en fonction de votre poste. Vous effectuerez une matrice de corrélation pour les attaquants (ST), les milieux de terrain (CM) et les défenseurs (CB) afin d'analyser les relations entre ces variables. 

#### D’après vos résultats, quels sont les joueurs les plus complets ? 


```python
attaquants = df[df['Position'] == 'ST']
milieux = df[df['Position'] == 'CM']
defenseurs = df[df['Position'] == 'CB']

skills = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'OVR']
          
plt.figure(figsize=(10, 6))
sns.heatmap(attaquants[skills].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between skills for ST')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(milieux[skills].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between skills for CM')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(defenseurs[skills].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between skills for CB')
plt.show()
```

Pour les attaquants, la note qui influence le plus l’OVR est : SHO = tir, PAC = vitesse et DRI = dribble.
Pour les milieux, la note qui influence le plus l’OVR est : PAS = passe, DRI = dribble, DEF = défense.
Pour les défenseurs, la note qui influence le plus l’OVR est : DEF = défense, PAC = vitesse.

Les points forts sont cohérents avec les rôles.
